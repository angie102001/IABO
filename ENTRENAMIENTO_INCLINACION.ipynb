```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd
wb=xlrd.open_workbook("/content/IABO/CARACTERIZACION_250_DATOS_DE_LA_MEDIA.xlsx")
sheet = wb.sheet_by_index(0)
#sheet.cell_value(0, 0)
ax=sheet.col_values(0)
ay=sheet.col_values(1)
az=sheet.col_values(2)
gx=sheet.col_values(3)
gy=sheet.col_values(4)
gz=sheet.col_values(5)
mx=sheet.col_values(6)
my=sheet.col_values(7)
mz=sheet.col_values(8)
clase = sheet.col_values(9)
plt.subplot(321)
plt.plot(ax)
plt.subplot(322)
plt.plot(ay)
plt.subplot(323)
plt.plot(az)
plt.subplot(324)
plt.plot(gx)
plt.subplot(325)
plt.plot(gy)
plt.subplot(326)
plt.plot(gz)
plt.subplot(324)
plt.plot(mx)
plt.subplot(325)
plt.plot(my)
plt.subplot(326)
plt.plot(mz)
```

```python
#plt.scatter(gx[0:50],gy[0:50],label = "DERECHA",color ="red")
#plt.scatter(gx[50:100],gy[50:100], label ="ABAJO", color = "blue")
#plt.scatter(gx[100:150],gy[100:150],label ="IZQUIERDA", color ="green")
#plt.scatter(gx[150:200],gy[150:200],label ="ESTABLE", color = "black")
#plt.scatter(gx[200:250],gy[200:250], label = "ARRIBA", color ="yellow")
#plt.legend()
def convertir_a_numero(dato):
  if dato == "DERECHA":
    return 0
  elif dato == "ABAJO":
    return 1
  elif dato == "IZQUIERDO":
    return 2
  elif dato == "ESTABLE":
    return 3
  else:
    return 4

  
claseNum = [convertir_a_numero(dato) for dato in clase ]
plt.scatter(gx,gy,c=claseNum)
plt.show()
```

```python
print(claseNum)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
matriz = np.array([gx,gy])
matriz_transpuesta = np.transpose(matriz)
matriz_transpuesta.shape
print(matriz_transpuesta)
```

```python
gx_train,gx_test,gy_train,gy_test=train_test_split(matriz_transpuesta,claseNum,test_size=0.2)
```

```python
vecinosTrain = KNeighborsClassifier(n_neighbors=4)
vecinosTrain.fit(gx_train,gy_train)
```

```python
clasificacion= vecinosTrain.predict(gx_test)
mat_con=confusion_matrix(gy_test,clasificacion)
print(mat_con)
print(accuracy_score(gy_test,clasificacion))
```

```python

nuevaflor = [[-0.00125,0.0001 ]] # Datos para probar
nueva = pd.DataFrame(nuevaflor) # Creacin de data frame de prueba
clasificacion= vecinosTrain.predict(nueva) # Uso del clasificador
print("Prediccin :")
if clasificacion == 0:
  print("DERECHA")
elif clasificacion == 1:
    print("ABAJO")
elif clasificacion == 2:
    print("IZQUIERDO")
elif clasificacion == 3:
    print("ESTABLE")
elif clasificacion == 4:
    print("ARRIBA")
```
